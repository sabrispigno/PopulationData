```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import pandas as pd
```

```python
df = pd.read_excel("PAS1.xlsx",sheet_name='ItSpSam')
df.set_index(df.columns[0], inplace=True)
dat=df.to_numpy()
```

```python
hist, bins= np.histogram(dat[799, :],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
bins_mid = (bins[1:]+bins[:-1])/2
plt.plot(bins_mid,hist)
plt.xscale('log')
```

```python
for i in range(dat.shape[0]):

    hist, bins= np.histogram(dat[i, :],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
    bins_mid = (bins[1:]+bins[:-1])/2
    plt.plot(bins_mid,hist)
    plt.xscale('log')
   
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import glob  # to find files

# Cerca tutti i file che corrispondono al pattern
file_pattern = "AbundancesDistributions/_nt1_simulation_beta=2.0_delta=*_S=50_mu=0.5std=0.3.npz"
files = glob.glob(file_pattern)
print(files[1])
```

```python
for file in files:
   
    delta = float(file.split("_delta=")[1].split("_")[0])
    
    # Carica i dati
    data = np.load(file)
    AFD = data['AFD']
    
    # Calcola l'istogramma
    hist, bins = np.histogram(AFD, bins=np.logspace(-10, 1, 50))
    bins_mid = (bins[1:] + bins[:-1]) / 2
    
    # Plotta l'istogramma come linea
    plt.plot(bins_mid, hist, label=f"delta={delta}")

plt.xscale('log')
plt.yscale('log')
plt.show()
```
