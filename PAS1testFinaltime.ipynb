```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import pandas as pd
```

```python
### Extract the dataset from xlsx
df = pd.read_excel("PAS1.xlsx",sheet_name='RAW')
df1=df[df["TimeSeries"]==4]
df2=df1.drop(df1.columns[[1, 2,3]], axis=1)
df2.set_index(df2.columns[0], inplace=True) # set the first row as index
df3=df2.transpose()
dat=df3.to_numpy()
```

```python
### Plotting single row Example
hist, bins= np.histogram(dat[799, :],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
bins_mid = (bins[1:]+bins[:-1])/2
plt.plot(bins_mid,hist)
plt.xscale('log')
```

```python
### Plotting all the rows (Species)
for i in range(dat.shape[0]):

    hist, bins= np.histogram(dat[i, :],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
    bins_mid = (bins[1:]+bins[:-1])/2
    plt.plot(bins_mid,hist)
    plt.xscale('log')
   
plt.show()

```

```python
### Plotting the mean of all the species 
mcdat=dat.mean(axis=0)
plt.hist(mcdat, bins=100)
plt.show
plt.xscale('log')
plt.yscale('log')
```

```python
### Plotting single column Example
hist, bins= np.histogram(dat[:, -1],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
bins_mid = (bins[1:]+bins[:-1])/2
plt.plot(bins_mid,hist)
plt.xscale('log')
```

```python
### Plotting all the columns (Samples)
for i in range(dat.shape[1]):

    hist, bins= np.histogram(dat[:, i],bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
    bins_mid = (bins[1:]+bins[:-1])/2
    plt.plot(bins_mid,hist)
    plt.xscale('log')
   
plt.show()
```

```python
### Plotting the mean of all the columns  
mcdat=dat.flatten()
hist, bins= np.histogram(mcdat,bins=np.logspace(-10,1,50)) # 50 bins in the range from 1e-10 e 1
bins_mid = (bins[1:]+bins[:-1])/2
plt.plot(bins_mid,hist)
plt.xscale('log')
plt.yscale('log')
```
