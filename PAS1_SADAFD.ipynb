```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import pandas as pd
```

```python
### Extract the dataset from xlsx
df = pd.read_excel("PAS1.xlsx",sheet_name='RAW')
df1=df[df["TimeSeries"]==8] #change to change the time
df2=df1.drop(df1.columns[[1, 2,3]], axis=1) # delate the unuseful columns
df2.set_index(df2.columns[0], inplace=True) # set the first row as index
df3=df2.transpose()
dat=df3.to_numpy()
### Plotting single column Example SAD Species Abundance Distribution 
hist, bins= np.histogram(dat[:, -1],bins=np.logspace(-10,1,50)) #x 50 bins in the range from 1e-10 e 1
bins_mid = (bins[1:]+bins[:-1])/2
plt.plot(bins_mid,hist)
plt.xscale('log')
```

```python
test=df1.to_numpy()
Diet= df1["dietType"]
Gender=df1["Gender"]

Diet=Diet.to_numpy()
Gender=Gender.to_numpy()

colors = {'CONTROL': 'orange', 'MED': 'green'}  
line_styles = {'Female': '-', 'Male': '--'}  
```

```python
### Plotting all the columns (Samples)
for i in range(dat.shape[1]):

    hist, bins= np.histogram(dat[:, i],bins=np.logspace(-8,1,50)) # 50 bins in the range from 1e-10 e 1
    bins_mid = (bins[1:]+bins[:-1])/2

    line_style = line_styles[Gender[i]]
    color = colors[Diet[i]]
    
    plt.plot(bins_mid,hist,line_style, color=color)
    plt.xscale('log')
    plt.ylim(0,50)
plt.show()
```
